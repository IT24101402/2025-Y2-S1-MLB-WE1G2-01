```python
# ==============================================================
# TEA LEAF DISEASE DETECTION — FINAL MLP MODEL (25 EPOCHS)
# Frozen MobileNetV2 + Deep MLP Classifier (with PCA, Plots & CV)
# ==============================================================

import os, glob, random, gc
import numpy as np
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt
import seaborn as sns

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models, regularizers
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint, Callback
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input as mobilenet_preprocess

from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA
from sklearn.metrics import (classification_report, confusion_matrix,
                             accuracy_score, precision_recall_fscore_support)
from sklearn.model_selection import StratifiedKFold
from sklearn.utils.class_weight import compute_class_weight

# ==============================================================
# STEP 1: SETUP
# ==============================================================
np.random.seed(42)
tf.random.set_seed(42)
random.seed(42)

DATA_ROOT = "/Users/tharukakumarasiri/Desktop/new model training/Final_preprocessed_Dataset"
TRAIN_DIR = os.path.join(DATA_ROOT, "train")
VAL_DIR   = os.path.join(DATA_ROOT, "val")
TEST_DIR  = os.path.join(DATA_ROOT, "test")

# ==============================================================
# STEP 2: LOAD IMAGE PATHS AND LABELS
# ==============================================================
def gather_paths_and_labels(root_dir):
    paths, labels = [], []
    for cls in sorted(os.listdir(root_dir)):
        cls_dir = os.path.join(root_dir, cls)
        if os.path.isdir(cls_dir):
            for p in glob.glob(os.path.join(cls_dir, "*.jpg")):
                paths.append(p)
                labels.append(cls)
    return paths, labels

class_names = sorted(os.listdir(TRAIN_DIR))
train_paths, train_labels = gather_paths_and_labels(TRAIN_DIR)
val_paths, val_labels     = gather_paths_and_labels(VAL_DIR)
test_paths, test_labels   = gather_paths_and_labels(TEST_DIR)

le = LabelEncoder().fit(class_names)
y_train = le.transform(train_labels)
y_val   = le.transform(val_labels)
y_test  = le.transform(test_labels)
NUM_CLASSES = len(class_names)

# ==============================================================
# STEP 3: LOAD AND PREPROCESS IMAGES
# ==============================================================
IMG_H, IMG_W = 224, 224

def load_images(paths):
    X = np.empty((len(paths), IMG_H, IMG_W, 3), dtype=np.float32)
    for i, path in enumerate(paths):
        img = Image.open(path).convert("RGB").resize((IMG_W, IMG_H))
        X[i] = np.asarray(img) / 255.0
    return X

X_train = load_images(train_paths)
X_val   = load_images(val_paths)
X_test  = load_images(test_paths)

# ==============================================================
# STEP 4: FEATURE EXTRACTION (Frozen MobileNetV2)
# ==============================================================
backbone = MobileNetV2(weights='imagenet', include_top=False,
                       pooling='avg', input_shape=(IMG_H, IMG_W, 3))
backbone.trainable = False

def extract_features(model, X, msg=""):
    X_prep = mobilenet_preprocess((X * 255).astype(np.float32))
    feats = model.predict(X_prep, batch_size=32, verbose=1)
    print(f"{msg} embeddings: {feats.shape}")
    return feats

X_train_emb = extract_features(backbone, X_train, "Train")
X_val_emb   = extract_features(backbone, X_val, "Val")
X_test_emb  = extract_features(backbone, X_test, "Test")
EMB_DIM = X_train_emb.shape[1]

# ==============================================================
# STEP 5: APPLY PCA (for dimensionality reduction)
# ==============================================================
pca = PCA(n_components=0.95, random_state=42)
X_train_pca = pca.fit_transform(X_train_emb)
X_val_pca   = pca.transform(X_val_emb)
X_test_pca  = pca.transform(X_test_emb)
print(f"PCA reduced dimensions: {X_train_pca.shape[1]}")

# ==============================================================
# STEP 6: CLASS WEIGHTS (to handle imbalance)
# ==============================================================
class_weights = dict(enumerate(
    compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
))

# ==============================================================
# STEP 7: BUILD THE MLP MODEL
# ==============================================================
def build_mlp(input_dim, num_classes):
    model = models.Sequential([
        layers.Input(shape=(input_dim,)),
        layers.Dense(1024, activation='relu', kernel_regularizer=regularizers.l2(1e-4)),
        layers.BatchNormalization(),
        layers.Dropout(0.5),
        layers.Dense(512, activation='relu', kernel_regularizer=regularizers.l2(1e-4)),
        layers.BatchNormalization(),
        layers.Dropout(0.4),
        layers.Dense(256, activation='relu'),
        layers.Dropout(0.3),
        layers.Dense(num_classes, activation='softmax')
    ])
    return model

mlp_model = build_mlp(X_train_pca.shape[1], NUM_CLASSES)
mlp_model.summary()

# ==============================================================
# STEP 8: COMPILE  (accuracy only — precision/recall via callback)
# ==============================================================
optimizer = keras.optimizers.Adam(learning_rate=5e-4)
loss_fn = keras.losses.SparseCategoricalCrossentropy()
mlp_model.compile(optimizer=optimizer, loss=loss_fn, metrics=["accuracy"])

# ==============================================================
# STEP 9: CALLBACKS
# ==============================================================
callbacks = [
    EarlyStopping(monitor="val_loss", patience=6, restore_best_weights=True),
    ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=3, min_lr=1e-6),
    ModelCheckpoint("best_final_mlp.keras", monitor="val_accuracy", save_best_only=True)
]

# ==============================================================
# STEP 10: PRECISION/RECALL CALLBACK (train + val, macro)
# ==============================================================
class EpochPR(Callback):
    def __init__(self, Xtr, ytr, Xva, yva, batch_size=32):
        super().__init__()
        self.Xtr, self.ytr, self.Xva, self.yva = Xtr, ytr, Xva, yva
        self.bs = batch_size
    def on_epoch_end(self, epoch, logs=None):
        # Train metrics
        p_tr = self.model.predict(self.Xtr, batch_size=self.bs, verbose=0)
        yhat_tr = np.argmax(p_tr, axis=1)
        pr_tr, rc_tr, _, _ = precision_recall_fscore_support(self.ytr, yhat_tr, average='macro', zero_division=0)
        # Val metrics
        p_va = self.model.predict(self.Xva, batch_size=self.bs, verbose=0)
        yhat_va = np.argmax(p_va, axis=1)
        pr_va, rc_va, _, _ = precision_recall_fscore_support(self.yva, yhat_va, average='macro', zero_division=0)

        # push into History so plots work
        h = self.model.history.history
        h.setdefault("precision", []).append(pr_tr)
        h.setdefault("val_precision", []).append(pr_va)
        h.setdefault("recall", []).append(rc_tr)
        h.setdefault("val_recall", []).append(rc_va)

        print(f" — epoch {epoch+1:02d}: precision {pr_tr:.4f}/{pr_va:.4f}  recall {rc_tr:.4f}/{rc_va:.4f}")

pr_cb = EpochPR(X_train_pca, y_train, X_val_pca, y_val, batch_size=32)

# ==============================================================
# STEP 11: TRAIN MODEL
# ==============================================================
history = mlp_model.fit(
    X_train_pca, y_train,
    validation_data=(X_val_pca, y_val),
    epochs=25,
    batch_size=32,
    class_weight=class_weights,
    callbacks=callbacks + [pr_cb],
    verbose=1
)

# ==============================================================
# STEP 12: TRAINING VISUALIZATION (Accuracy/Loss/Precision/Recall)
# ==============================================================
def plot_training(history):
    h = history.history
    fig, axes = plt.subplots(2, 2, figsize=(14,10))
    fig.suptitle("Training History — Final MLP Model", fontsize=16, fontweight='bold')

    axes[0,0].plot(h["accuracy"], label="Train Acc")
    axes[0,0].plot(h["val_accuracy"], label="Val Acc")
    axes[0,0].set_title("Accuracy"); axes[0,0].legend(); axes[0,0].grid(alpha=0.3)

    axes[0,1].plot(h["loss"], label="Train Loss")
    axes[0,1].plot(h["val_loss"], label="Val Loss")
    axes[0,1].set_title("Loss"); axes[0,1].legend(); axes[0,1].grid(alpha=0.3)

    axes[1,0].plot(h.get("precision", []), label="Train Precision")
    axes[1,0].plot(h.get("val_precision", []), label="Val Precision")
    axes[1,0].set_title("Precision"); axes[1,0].legend(); axes[1,0].grid(alpha=0.3)

    axes[1,1].plot(h.get("recall", []), label="Train Recall")
    axes[1,1].plot(h.get("val_recall", []), label="Val Recall")
    axes[1,1].set_title("Recall"); axes[1,1].legend(); axes[1,1].grid(alpha=0.3)

    plt.tight_layout(); plt.show()

plot_training(history)

# ==============================================================
# STEP 13: EVALUATION & REPORT
# ==============================================================
best_model = keras.models.load_model("best_final_mlp.keras")
y_pred = np.argmax(best_model.predict(X_test_pca), axis=1)

acc = accuracy_score(y_test, y_pred)
prec, rec, f1, _ = precision_recall_fscore_support(y_test, y_pred, average="macro")
print(f"Accuracy: {acc:.4f} | Precision: {prec:.4f} | Recall: {rec:.4f} | F1: {f1:.4f}")

cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(9,7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.title("Confusion Matrix — Final MLP Model")
plt.xlabel("Predicted"); plt.ylabel("True"); plt.tight_layout(); plt.show()

print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=class_names))

# ==============================================================
# STEP 14: CROSS-VALIDATION
# ==============================================================
print("\nPerforming 5-Fold Cross Validation...")
X_full = np.vstack([X_train_pca, X_val_pca])
y_full = np.concatenate([y_train, y_val])
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = []

for train_idx, val_idx in skf.split(X_full, y_full):
    model_cv = build_mlp(X_full.shape[1], NUM_CLASSES)
    model_cv.compile(optimizer=keras.optimizers.Adam(learning_rate=5e-4),
                     loss=loss_fn, metrics=["accuracy"])
    model_cv.fit(X_full[train_idx], y_full[train_idx],
                 validation_data=(X_full[val_idx], y_full[val_idx]),
                 epochs=10, batch_size=32, verbose=0)
    val_acc = model_cv.evaluate(X_full[val_idx], y_full[val_idx], verbose=0)[1]
    cv_scores.append(val_acc)

print(f"Cross-Validation Mean Accuracy: {np.mean(cv_scores):.4f} ± {np.std(cv_scores):.4f}")

# ==============================================================
# STEP 15: SAVE RESULTS
# ==============================================================
results = pd.DataFrame({
    "Model": ["Final_MLP_MobileNetV2"],
    "Accuracy": [acc],
    "Precision": [prec],
    "Recall": [rec],
    "F1_Score": [f1],
    "CV_Mean_Acc": [np.mean(cv_scores)],
    "CV_Std": [np.std(cv_scores)]
})
results.to_csv("mlp_final_results.csv", index=False)
print("\n✅ Results saved as 'mlp_final_results.csv'")
```

```python
# ==============================================================
# TEA LEAF DISEASE DETECTION — HYPERPARAMETER-TUNED MLP MODEL
# MobileNetV2 (Frozen) + Deep LeakyReLU MLP Classifier
# Epochs = 50 | SGD | Higher Dropout | Larger Layers | Cross-Validation + CSV
# ==============================================================

import os, glob, random, gc
import numpy as np
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt
import seaborn as sns

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models, regularizers
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint, Callback
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input as mobilenet_preprocess

from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA
from sklearn.metrics import (classification_report, confusion_matrix,
                             accuracy_score, precision_recall_fscore_support)
from sklearn.model_selection import StratifiedKFold
from sklearn.utils.class_weight import compute_class_weight

# ==============================================================
# STEP 1: SETUP
# ==============================================================
np.random.seed(42)
tf.random.set_seed(42)
random.seed(42)

DATA_ROOT = "/Users/tharukakumarasiri/Desktop/new model training/Final_preprocessed_Dataset"
TRAIN_DIR = os.path.join(DATA_ROOT, "train")
VAL_DIR   = os.path.join(DATA_ROOT, "val")
TEST_DIR  = os.path.join(DATA_ROOT, "test")

# ==============================================================
# STEP 2: LOAD PATHS + LABELS
# ==============================================================
def gather_paths_and_labels(root_dir):
    paths, labels = [], []
    for cls in sorted(os.listdir(root_dir)):
        cls_dir = os.path.join(root_dir, cls)
        if os.path.isdir(cls_dir):
            for p in glob.glob(os.path.join(cls_dir, "*.jpg")):
                paths.append(p)
                labels.append(cls)
    return paths, labels

class_names = sorted(os.listdir(TRAIN_DIR))
train_paths, train_labels = gather_paths_and_labels(TRAIN_DIR)
val_paths, val_labels     = gather_paths_and_labels(VAL_DIR)
test_paths, test_labels   = gather_paths_and_labels(TEST_DIR)

le = LabelEncoder().fit(class_names)
y_train = le.transform(train_labels)
y_val   = le.transform(val_labels)
y_test  = le.transform(test_labels)
NUM_CLASSES = len(class_names)

# ==============================================================
# STEP 3: LOAD IMAGES
# ==============================================================
IMG_H, IMG_W = 224, 224

def load_images(paths):
    X = np.empty((len(paths), IMG_H, IMG_W, 3), dtype=np.float32)
    for i, path in enumerate(paths):
        img = Image.open(path).convert("RGB").resize((IMG_W, IMG_H))
        X[i] = np.asarray(img) / 255.0
    return X

X_train = load_images(train_paths)
X_val   = load_images(val_paths)
X_test  = load_images(test_paths)

# ==============================================================
# STEP 4: FEATURE EXTRACTION (MobileNetV2 Frozen)
# ==============================================================
backbone = MobileNetV2(weights='imagenet', include_top=False,
                       pooling='avg', input_shape=(IMG_H, IMG_W, 3))
backbone.trainable = False

def extract_features(model, X, msg=""):
    X_prep = mobilenet_preprocess((X * 255).astype(np.float32))
    feats = model.predict(X_prep, batch_size=32, verbose=1)
    print(f"{msg} embeddings: {feats.shape}")
    return feats

X_train_emb = extract_features(backbone, X_train, "Train")
X_val_emb   = extract_features(backbone, X_val, "Val")
X_test_emb  = extract_features(backbone, X_test, "Test")

# ==============================================================
# STEP 5: PCA
# ==============================================================
pca = PCA(n_components=0.95, random_state=42)
X_train_pca = pca.fit_transform(X_train_emb)
X_val_pca   = pca.transform(X_val_emb)
X_test_pca  = pca.transform(X_test_emb)
print(f"PCA reduced dim: {X_train_pca.shape[1]}")

# ==============================================================
# STEP 6: CLASS WEIGHTS
# ==============================================================
class_weights = dict(enumerate(
    compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
))

# ==============================================================
# STEP 7: BUILD MODEL
# ==============================================================
def build_tuned_mlp(input_dim, num_classes):
    model = models.Sequential([
        layers.Input(shape=(input_dim,)),
        layers.Dense(2048, kernel_regularizer=regularizers.l2(1e-4)),
        layers.BatchNormalization(),
        layers.LeakyReLU(alpha=0.1),
        layers.Dropout(0.6),

        layers.Dense(1024, kernel_regularizer=regularizers.l2(1e-4)),
        layers.BatchNormalization(),
        layers.LeakyReLU(alpha=0.1),
        layers.Dropout(0.5),

        layers.Dense(512, kernel_regularizer=regularizers.l2(1e-4)),
        layers.BatchNormalization(),
        layers.LeakyReLU(alpha=0.1),
        layers.Dropout(0.4),

        layers.Dense(256, kernel_regularizer=regularizers.l2(1e-4)),
        layers.BatchNormalization(),
        layers.LeakyReLU(alpha=0.1),
        layers.Dropout(0.3),

        layers.Dense(num_classes, activation='softmax')
    ])
    return model

# ==============================================================
# STEP 8: COMPILE
# ==============================================================
optimizer = keras.optimizers.SGD(learning_rate=0.01, momentum=0.9, nesterov=True)
loss_fn = keras.losses.SparseCategoricalCrossentropy()
tuned_model = build_tuned_mlp(X_train_pca.shape[1], NUM_CLASSES)
tuned_model.compile(optimizer=optimizer, loss=loss_fn, metrics=["accuracy"])

# ==============================================================
# STEP 9: CALLBACKS
# ==============================================================
callbacks = [
    EarlyStopping(monitor="val_loss", patience=8, restore_best_weights=True),
    ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=4, min_lr=1e-6),
    ModelCheckpoint("best_tuned_mlp.keras", monitor="val_accuracy", save_best_only=True)
]

# ==============================================================
# STEP 10: PRECISION/RECALL CALLBACK
# ==============================================================
class EpochPR(Callback):
    def __init__(self, Xtr, ytr, Xva, yva, batch_size=32):
        super().__init__()
        self.Xtr, self.ytr, self.Xva, self.yva = Xtr, ytr, Xva, yva
        self.bs = batch_size
    def on_epoch_end(self, epoch, logs=None):
        p_tr = self.model.predict(self.Xtr, batch_size=self.bs, verbose=0)
        p_va = self.model.predict(self.Xva, batch_size=self.bs, verbose=0)
        yhat_tr = np.argmax(p_tr, axis=1)
        yhat_va = np.argmax(p_va, axis=1)
        pr_tr, rc_tr, _, _ = precision_recall_fscore_support(self.ytr, yhat_tr, average='macro', zero_division=0)
        pr_va, rc_va, _, _ = precision_recall_fscore_support(self.yva, yhat_va, average='macro', zero_division=0)
        h = self.model.history.history
        h.setdefault("precision", []).append(pr_tr)
        h.setdefault("val_precision", []).append(pr_va)
        h.setdefault("recall", []).append(rc_tr)
        h.setdefault("val_recall", []).append(rc_va)
        print(f"Epoch {epoch+1}: Precision {pr_tr:.4f}/{pr_va:.4f} Recall {rc_tr:.4f}/{rc_va:.4f}")

pr_cb = EpochPR(X_train_pca, y_train, X_val_pca, y_val)

# ==============================================================
# STEP 11: TRAIN MODEL
# ==============================================================
history = tuned_model.fit(
    X_train_pca, y_train,
    validation_data=(X_val_pca, y_val),
    epochs=50,
    batch_size=32,
    class_weight=class_weights,
    callbacks=callbacks + [pr_cb],
    verbose=1
)

# ==============================================================
# STEP 12: TRAINING PLOTS
# ==============================================================
def plot_training(history):
    h = history.history
    fig, axes = plt.subplots(2, 2, figsize=(14,10))
    fig.suptitle("Training History — Tuned MLP Model (SGD + LeakyReLU)", fontsize=16, fontweight='bold')

    axes[0,0].plot(h["accuracy"], label="Train Acc")
    axes[0,0].plot(h["val_accuracy"], label="Val Acc")
    axes[0,0].set_title("Accuracy"); axes[0,0].legend(); axes[0,0].grid(alpha=0.3)

    axes[0,1].plot(h["loss"], label="Train Loss")
    axes[0,1].plot(h["val_loss"], label="Val Loss")
    axes[0,1].set_title("Loss"); axes[0,1].legend(); axes[0,1].grid(alpha=0.3)

    axes[1,0].plot(h.get("precision", []), label="Train Precision")
    axes[1,0].plot(h.get("val_precision", []), label="Val Precision")
    axes[1,0].set_title("Precision"); axes[1,0].legend(); axes[1,0].grid(alpha=0.3)

    axes[1,1].plot(h.get("recall", []), label="Train Recall")
    axes[1,1].plot(h.get("val_recall", []), label="Val Recall")
    axes[1,1].set_title("Recall"); axes[1,1].legend(); axes[1,1].grid(alpha=0.3)

    plt.tight_layout(); plt.show()

plot_training(history)

# ==============================================================
# STEP 13: EVALUATION & REPORT
# ==============================================================
best_model = keras.models.load_model("best_tuned_mlp.keras")
y_pred = np.argmax(best_model.predict(X_test_pca), axis=1)

acc = accuracy_score(y_test, y_pred)
prec, rec, f1, _ = precision_recall_fscore_support(y_test, y_pred, average="macro")

print(f"\nFinal Test Performance → Accuracy: {acc:.4f} | Precision: {prec:.4f} | Recall: {rec:.4f} | F1: {f1:.4f}")

cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(9,7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Greens',
            xticklabels=class_names, yticklabels=class_names)
plt.title("Confusion Matrix — Tuned MLP Model (SGD + LeakyReLU)")
plt.xlabel("Predicted"); plt.ylabel("True"); plt.tight_layout(); plt.show()

print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=class_names))

# ==============================================================
# STEP 14: CROSS-VALIDATION
# ==============================================================
print("\nPerforming 5-Fold Cross-Validation...")
X_full = np.vstack([X_train_pca, X_val_pca])
y_full = np.concatenate([y_train, y_val])
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = []

for fold, (tr_idx, va_idx) in enumerate(skf.split(X_full, y_full), 1):
    print(f"\nFold {fold}")
    model_cv = build_tuned_mlp(X_full.shape[1], NUM_CLASSES)
    model_cv.compile(optimizer=keras.optimizers.SGD(learning_rate=0.01, momentum=0.9, nesterov=True),
                     loss=loss_fn, metrics=["accuracy"])
    model_cv.fit(X_full[tr_idx], y_full[tr_idx],
                 validation_data=(X_full[va_idx], y_full[va_idx]),
                 epochs=10, batch_size=32, verbose=0)
    val_acc = model_cv.evaluate(X_full[va_idx], y_full[va_idx], verbose=0)[1]
    print(f" → Fold {fold} Accuracy: {val_acc:.4f}")
    cv_scores.append(val_acc)

cv_mean, cv_std = np.mean(cv_scores), np.std(cv_scores)
print(f"\nCross-Validation Mean Accuracy: {cv_mean:.4f} ± {cv_std:.4f}")

# ==============================================================
# STEP 15: SAVE RESULTS AS CSV
# ==============================================================
results = pd.DataFrame({
    "Model": ["Tuned_MLP_MobileNetV2_SGD"],
    "Accuracy": [acc],
    "Precision": [prec],
    "Recall": [rec],
    "F1_Score": [f1],
    "CV_Mean_Acc": [cv_mean],
    "CV_Std": [cv_std]
})
results.to_csv("tuned_mlp_results.csv", index=False)
print("\n✅ Results saved as 'tuned_mlp_results.csv'")
```

```python
# ==============================================================
# TEA LEAF DISEASE DETECTION — HYPERPARAMETER-TUNED MLP MODEL
# MobileNetV2 (Frozen) + Deep LeakyReLU MLP Classifier
# Epochs = 50 | SGD | Higher Dropout | Larger Layers | Cross-Validation + CSV
# ==============================================================

import os, glob, random, gc
import numpy as np
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt
import seaborn as sns

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models, regularizers
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint, Callback
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input as mobilenet_preprocess

from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA
from sklearn.metrics import (classification_report, confusion_matrix,
                             accuracy_score, precision_recall_fscore_support)
from sklearn.model_selection import StratifiedKFold
from sklearn.utils.class_weight import compute_class_weight

# ==============================================================
# STEP 1: SETUP
# ==============================================================
np.random.seed(42)
tf.random.set_seed(42)
random.seed(42)

DATA_ROOT = "/Users/tharukakumarasiri/Desktop/new model training/Final_preprocessed_Dataset"
TRAIN_DIR = os.path.join(DATA_ROOT, "train")
VAL_DIR   = os.path.join(DATA_ROOT, "val")
TEST_DIR  = os.path.join(DATA_ROOT, "test")

# ==============================================================
# STEP 2: LOAD PATHS + LABELS
# ==============================================================
def gather_paths_and_labels(root_dir):
    paths, labels = [], []
    for cls in sorted(os.listdir(root_dir)):
        cls_dir = os.path.join(root_dir, cls)
        if os.path.isdir(cls_dir):
            for p in glob.glob(os.path.join(cls_dir, "*.jpg")):
                paths.append(p)
                labels.append(cls)
    return paths, labels

class_names = sorted(os.listdir(TRAIN_DIR))
train_paths, train_labels = gather_paths_and_labels(TRAIN_DIR)
val_paths, val_labels     = gather_paths_and_labels(VAL_DIR)
test_paths, test_labels   = gather_paths_and_labels(TEST_DIR)

le = LabelEncoder().fit(class_names)
y_train = le.transform(train_labels)
y_val   = le.transform(val_labels)
y_test  = le.transform(test_labels)
NUM_CLASSES = len(class_names)

# ==============================================================
# STEP 3: LOAD IMAGES
# ==============================================================
IMG_H, IMG_W = 224, 224

def load_images(paths):
    X = np.empty((len(paths), IMG_H, IMG_W, 3), dtype=np.float32)
    for i, path in enumerate(paths):
        img = Image.open(path).convert("RGB").resize((IMG_W, IMG_H))
        X[i] = np.asarray(img) / 255.0
    return X

X_train = load_images(train_paths)
X_val   = load_images(val_paths)
X_test  = load_images(test_paths)

# ==============================================================
# STEP 4: FEATURE EXTRACTION (MobileNetV2 Frozen)
# ==============================================================
backbone = MobileNetV2(weights='imagenet', include_top=False,
                       pooling='avg', input_shape=(IMG_H, IMG_W, 3))
backbone.trainable = False

def extract_features(model, X, msg=""):
    X_prep = mobilenet_preprocess((X * 255).astype(np.float32))
    feats = model.predict(X_prep, batch_size=32, verbose=1)
    print(f"{msg} embeddings: {feats.shape}")
    return feats

X_train_emb = extract_features(backbone, X_train, "Train")
X_val_emb   = extract_features(backbone, X_val, "Val")
X_test_emb  = extract_features(backbone, X_test, "Test")

# ==============================================================
# STEP 5: PCA
# ==============================================================
pca = PCA(n_components=0.95, random_state=42)
X_train_pca = pca.fit_transform(X_train_emb)
X_val_pca   = pca.transform(X_val_emb)
X_test_pca  = pca.transform(X_test_emb)
print(f"PCA reduced dim: {X_train_pca.shape[1]}")

# ==============================================================
# STEP 6: CLASS WEIGHTS
# ==============================================================
class_weights = dict(enumerate(
    compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
))

# ==============================================================
# STEP 7: BUILD MODEL
# ==============================================================
def build_tuned_mlp(input_dim, num_classes):
    model = models.Sequential([
        layers.Input(shape=(input_dim,)),
        layers.Dense(2048, kernel_regularizer=regularizers.l2(1e-4)),
        layers.BatchNormalization(),
        layers.LeakyReLU(alpha=0.1),
        layers.Dropout(0.6),

        layers.Dense(1024, kernel_regularizer=regularizers.l2(1e-4)),
        layers.BatchNormalization(),
        layers.LeakyReLU(alpha=0.1),
        layers.Dropout(0.5),

        layers.Dense(512, kernel_regularizer=regularizers.l2(1e-4)),
        layers.BatchNormalization(),
        layers.LeakyReLU(alpha=0.1),
        layers.Dropout(0.4),

        layers.Dense(256, kernel_regularizer=regularizers.l2(1e-4)),
        layers.BatchNormalization(),
        layers.LeakyReLU(alpha=0.1),
        layers.Dropout(0.3),

        layers.Dense(num_classes, activation='softmax')
    ])
    return model

# ==============================================================
# STEP 8: COMPILE
# ==============================================================
optimizer = keras.optimizers.SGD(learning_rate=0.01, momentum=0.9, nesterov=True)
loss_fn = keras.losses.SparseCategoricalCrossentropy()
tuned_model = build_tuned_mlp(X_train_pca.shape[1], NUM_CLASSES)
tuned_model.compile(optimizer=optimizer, loss=loss_fn, metrics=["accuracy"])

# ==============================================================
# STEP 9: CALLBACKS
# ==============================================================
callbacks = [
    EarlyStopping(monitor="val_loss", patience=8, restore_best_weights=True),
    ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=4, min_lr=1e-6),
    ModelCheckpoint("best_tuned_mlp.keras", monitor="val_accuracy", save_best_only=True)
]

# ==============================================================
# STEP 10: PRECISION/RECALL CALLBACK
# ==============================================================
class EpochPR(Callback):
    def __init__(self, Xtr, ytr, Xva, yva, batch_size=32):
        super().__init__()
        self.Xtr, self.ytr, self.Xva, self.yva = Xtr, ytr, Xva, yva
        self.bs = batch_size
    def on_epoch_end(self, epoch, logs=None):
        p_tr = self.model.predict(self.Xtr, batch_size=self.bs, verbose=0)
        p_va = self.model.predict(self.Xva, batch_size=self.bs, verbose=0)
        yhat_tr = np.argmax(p_tr, axis=1)
        yhat_va = np.argmax(p_va, axis=1)
        pr_tr, rc_tr, _, _ = precision_recall_fscore_support(self.ytr, yhat_tr, average='macro', zero_division=0)
        pr_va, rc_va, _, _ = precision_recall_fscore_support(self.yva, yhat_va, average='macro', zero_division=0)
        h = self.model.history.history
        h.setdefault("precision", []).append(pr_tr)
        h.setdefault("val_precision", []).append(pr_va)
        h.setdefault("recall", []).append(rc_tr)
        h.setdefault("val_recall", []).append(rc_va)
        print(f"Epoch {epoch+1}: Precision {pr_tr:.4f}/{pr_va:.4f} Recall {rc_tr:.4f}/{rc_va:.4f}")

pr_cb = EpochPR(X_train_pca, y_train, X_val_pca, y_val)

# ==============================================================
# STEP 11: TRAIN MODEL
# ==============================================================
history = tuned_model.fit(
    X_train_pca, y_train,
    validation_data=(X_val_pca, y_val),
    epochs=75,
    batch_size=32,
    class_weight=class_weights,
    callbacks=callbacks + [pr_cb],
    verbose=1
)

# ==============================================================
# STEP 12: TRAINING PLOTS
# ==============================================================
def plot_training(history):
    h = history.history
    fig, axes = plt.subplots(2, 2, figsize=(14,10))
    fig.suptitle("Training History — Tuned MLP Model (SGD + LeakyReLU)", fontsize=16, fontweight='bold')

    axes[0,0].plot(h["accuracy"], label="Train Acc")
    axes[0,0].plot(h["val_accuracy"], label="Val Acc")
    axes[0,0].set_title("Accuracy"); axes[0,0].legend(); axes[0,0].grid(alpha=0.3)

    axes[0,1].plot(h["loss"], label="Train Loss")
    axes[0,1].plot(h["val_loss"], label="Val Loss")
    axes[0,1].set_title("Loss"); axes[0,1].legend(); axes[0,1].grid(alpha=0.3)

    axes[1,0].plot(h.get("precision", []), label="Train Precision")
    axes[1,0].plot(h.get("val_precision", []), label="Val Precision")
    axes[1,0].set_title("Precision"); axes[1,0].legend(); axes[1,0].grid(alpha=0.3)

    axes[1,1].plot(h.get("recall", []), label="Train Recall")
    axes[1,1].plot(h.get("val_recall", []), label="Val Recall")
    axes[1,1].set_title("Recall"); axes[1,1].legend(); axes[1,1].grid(alpha=0.3)

    plt.tight_layout(); plt.show()

plot_training(history)

# ==============================================================
# STEP 13: EVALUATION & REPORT
# ==============================================================
best_model = keras.models.load_model("best_tuned_mlp.keras")
y_pred = np.argmax(best_model.predict(X_test_pca), axis=1)

acc = accuracy_score(y_test, y_pred)
prec, rec, f1, _ = precision_recall_fscore_support(y_test, y_pred, average="macro")

print(f"\nFinal Test Performance → Accuracy: {acc:.4f} | Precision: {prec:.4f} | Recall: {rec:.4f} | F1: {f1:.4f}")

cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(9,7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Greens',
            xticklabels=class_names, yticklabels=class_names)
plt.title("Confusion Matrix — Tuned MLP Model (SGD + LeakyReLU)")
plt.xlabel("Predicted"); plt.ylabel("True"); plt.tight_layout(); plt.show()

print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=class_names))

# ==============================================================
# STEP 14: CROSS-VALIDATION
# ==============================================================
print("\nPerforming 5-Fold Cross-Validation...")
X_full = np.vstack([X_train_pca, X_val_pca])
y_full = np.concatenate([y_train, y_val])
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = []

for fold, (tr_idx, va_idx) in enumerate(skf.split(X_full, y_full), 1):
    print(f"\nFold {fold}")
    model_cv = build_tuned_mlp(X_full.shape[1], NUM_CLASSES)
    model_cv.compile(optimizer=keras.optimizers.SGD(learning_rate=0.01, momentum=0.9, nesterov=True),
                     loss=loss_fn, metrics=["accuracy"])
    model_cv.fit(X_full[tr_idx], y_full[tr_idx],
                 validation_data=(X_full[va_idx], y_full[va_idx]),
                 epochs=10, batch_size=32, verbose=0)
    val_acc = model_cv.evaluate(X_full[va_idx], y_full[va_idx], verbose=0)[1]
    print(f" → Fold {fold} Accuracy: {val_acc:.4f}")
    cv_scores.append(val_acc)

cv_mean, cv_std = np.mean(cv_scores), np.std(cv_scores)
print(f"\nCross-Validation Mean Accuracy: {cv_mean:.4f} ± {cv_std:.4f}")

# ==============================================================
# STEP 15: SAVE RESULTS AS CSV
# ==============================================================
results = pd.DataFrame({
    "Model": ["Tuned_MLP_MobileNetV2_SGD"],
    "Accuracy": [acc],
    "Precision": [prec],
    "Recall": [rec],
    "F1_Score": [f1],
    "CV_Mean_Acc": [cv_mean],
    "CV_Std": [cv_std]
})
results.to_csv("tuned_mlp_results2.csv", index=False)
print("\n✅ Results saved as 'tuned_mlp_results.csv'")
```

```python
# ==============================================================
# TEA LEAF DISEASE DETECTION — FINAL MODERN MODEL
# MobileNetV2 (Partially Fine-Tuned) + Deep LeakyReLU MLP Classifier
# One-Hot Labels + Label Smoothing + GaussianNoise + SGD + LR Decay
# ==============================================================

import os, glob, random, gc
import numpy as np
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt
import seaborn as sns

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models, regularizers
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint, Callback, LearningRateScheduler
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input as mobilenet_preprocess
from tensorflow.keras.utils import to_categorical

from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA
from sklearn.metrics import (classification_report, confusion_matrix,
                             accuracy_score, precision_recall_fscore_support)
from sklearn.utils.class_weight import compute_class_weight

# ==============================================================
# STEP 1: SETUP
# ==============================================================
np.random.seed(42)
tf.random.set_seed(42)
random.seed(42)

DATA_ROOT = "/Users/tharukakumarasiri/Desktop/new model training/Final_preprocessed_Dataset"
TRAIN_DIR = os.path.join(DATA_ROOT, "train")
VAL_DIR   = os.path.join(DATA_ROOT, "val")
TEST_DIR  = os.path.join(DATA_ROOT, "test")

# ==============================================================
# STEP 2: LOAD PATHS + LABELS
# ==============================================================
def gather_paths_and_labels(root_dir):
    paths, labels = [], []
    for cls in sorted(os.listdir(root_dir)):
        cls_dir = os.path.join(root_dir, cls)
        if os.path.isdir(cls_dir):
            for p in glob.glob(os.path.join(cls_dir, "*.jpg")):
                paths.append(p)
                labels.append(cls)
    return paths, labels

class_names = sorted(os.listdir(TRAIN_DIR))
train_paths, train_labels = gather_paths_and_labels(TRAIN_DIR)
val_paths, val_labels     = gather_paths_and_labels(VAL_DIR)
test_paths, test_labels   = gather_paths_and_labels(TEST_DIR)

le = LabelEncoder().fit(class_names)
y_train = le.transform(train_labels)
y_val   = le.transform(val_labels)
y_test  = le.transform(test_labels)
NUM_CLASSES = len(class_names)

# ✅ Convert to One-Hot Encoded format for CategoricalCrossentropy
y_train = to_categorical(y_train, NUM_CLASSES)
y_val   = to_categorical(y_val, NUM_CLASSES)
y_test  = to_categorical(y_test, NUM_CLASSES)

# ==============================================================
# STEP 3: LOAD IMAGES
# ==============================================================
IMG_H, IMG_W = 224, 224

def load_images(paths):
    X = np.empty((len(paths), IMG_H, IMG_W, 3), dtype=np.float32)
    for i, path in enumerate(paths):
        img = Image.open(path).convert("RGB").resize((IMG_W, IMG_H))
        X[i] = np.asarray(img) / 255.0
    return X

X_train = load_images(train_paths)
X_val   = load_images(val_paths)
X_test  = load_images(test_paths)

# ==============================================================
# STEP 4: DATA AUGMENTATION
# ==============================================================
datagen = tf.keras.preprocessing.image.ImageDataGenerator(
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    fill_mode="nearest"
)

# ==============================================================
# STEP 5: FEATURE EXTRACTION (Fine-Tuned MobileNetV2)
# ==============================================================
backbone = MobileNetV2(weights='imagenet', include_top=False,
                       pooling='avg', input_shape=(IMG_H, IMG_W, 3))

for layer in backbone.layers[:-30]:
    layer.trainable = False
for layer in backbone.layers[-30:]:
    layer.trainable = True

def extract_features(model, X, msg=""):
    X_prep = mobilenet_preprocess((X * 255).astype(np.float32))
    feats = model.predict(X_prep, batch_size=32, verbose=1)
    print(f"{msg} embeddings: {feats.shape}")
    return feats

X_train_emb = extract_features(backbone, X_train, "Train")
X_val_emb   = extract_features(backbone, X_val, "Val")
X_test_emb  = extract_features(backbone, X_test, "Test")

# ==============================================================
# STEP 6: PCA
# ==============================================================
pca = PCA(n_components=0.95, random_state=42)
X_train_pca = pca.fit_transform(X_train_emb)
X_val_pca   = pca.transform(X_val_emb)
X_test_pca  = pca.transform(X_test_emb)
print(f"PCA reduced dim: {X_train_pca.shape[1]}")

# ==============================================================
# STEP 7: BUILD MODEL
# ==============================================================
def build_final_mlp(input_dim, num_classes):
    model = models.Sequential([
        layers.Input(shape=(input_dim,)),
        layers.GaussianNoise(0.05),

        layers.Dense(2048, kernel_regularizer=regularizers.l2(1e-3)),
        layers.BatchNormalization(),
        layers.LeakyReLU(alpha=0.1),
        layers.Dropout(0.5),

        layers.Dense(1024, kernel_regularizer=regularizers.l2(1e-3)),
        layers.BatchNormalization(),
        layers.LeakyReLU(alpha=0.1),
        layers.Dropout(0.45),

        layers.Dense(512, kernel_regularizer=regularizers.l2(1e-3)),
        layers.BatchNormalization(),
        layers.LeakyReLU(alpha=0.1),
        layers.Dropout(0.4),

        layers.Dense(256, kernel_regularizer=regularizers.l2(1e-3)),
        layers.BatchNormalization(),
        layers.LeakyReLU(alpha=0.1),
        layers.Dropout(0.4),

        layers.Dense(num_classes, activation='softmax')
    ])
    return model

# ==============================================================
# STEP 8: COMPILE (Modern — Label Smoothing)
# ==============================================================
initial_lr = 0.01
optimizer = keras.optimizers.SGD(learning_rate=initial_lr, momentum=0.9, nesterov=True)

def lr_decay(epoch, lr):
    decay_rate = 0.9
    if epoch % 5 == 0 and epoch > 0:
        lr *= decay_rate
    return lr

lr_scheduler = LearningRateScheduler(lr_decay)

# ✅ Modern loss with label smoothing (requires one-hot labels)
loss_fn = keras.losses.CategoricalCrossentropy(label_smoothing=0.1)

final_model = build_final_mlp(X_train_pca.shape[1], NUM_CLASSES)
final_model.compile(optimizer=optimizer, loss=loss_fn, metrics=["accuracy"])

# ==============================================================
# STEP 9: CALLBACKS
# ==============================================================
callbacks = [
    EarlyStopping(monitor="val_loss", patience=10, restore_best_weights=True),
    ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=3, min_lr=1e-6),
    ModelCheckpoint("best_final_model.keras", monitor="val_accuracy", save_best_only=True),
    lr_scheduler
]

# ==============================================================
# STEP 10: TRAIN
# ==============================================================
history = final_model.fit(
    X_train_pca, y_train,
    validation_data=(X_val_pca, y_val),
    epochs=50,
    batch_size=32,
    callbacks=callbacks,
    verbose=1
)

# ==============================================================
# STEP 11: PLOTS
# ==============================================================
def plot_training(history):
    h = history.history
    plt.figure(figsize=(12,5))
    plt.subplot(1,2,1)
    plt.plot(h["accuracy"], label="Train Acc")
    plt.plot(h["val_accuracy"], label="Val Acc")
    plt.legend(); plt.title("Accuracy"); plt.grid(alpha=0.3)
    plt.subplot(1,2,2)
    plt.plot(h["loss"], label="Train Loss")
    plt.plot(h["val_loss"], label="Val Loss")
    plt.legend(); plt.title("Loss"); plt.grid(alpha=0.3)
    plt.tight_layout(); plt.show()

plot_training(history)

# ==============================================================
# STEP 12: EVALUATE & SAVE
# ==============================================================
best_model = keras.models.load_model("best_final_model.keras")
y_pred = np.argmax(best_model.predict(X_test_pca), axis=1)
y_true = np.argmax(y_test, axis=1)

acc = accuracy_score(y_true, y_pred)
prec, rec, f1, _ = precision_recall_fscore_support(y_true, y_pred, average="macro")

print(f"\nFinal Test Performance → Accuracy: {acc:.4f} | Precision: {prec:.4f} | Recall: {rec:.4f} | F1: {f1:.4f}")

cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(9,7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Greens',
            xticklabels=class_names, yticklabels=class_names)
plt.title("Confusion Matrix — Final Modern Model")
plt.xlabel("Predicted"); plt.ylabel("True")
plt.tight_layout(); plt.show()

print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=class_names))

# Save results to CSV
results = pd.DataFrame({
    "Model": ["Final_Modern_MLP_MobileNetV2_SGD_FT"],
    "Accuracy": [acc],
    "Precision": [prec],
    "Recall": [rec],
    "F1_Score": [f1],
})
results.to_csv("final_modern_results.csv", index=False)
print("\n✅ Results saved as 'final_modern_results.csv'")
```
